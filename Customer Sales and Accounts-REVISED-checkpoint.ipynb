```python
#concatenate first name and last name columns to create customer_name to create common data point
cactdf['CUSTOMER_NAME'] = cactdf['CUSTOMER_FIRST_NAME'].map(str) + ' ' + cactdf['CUSTOMER_LAST_NAME'].map(str)
#Confirm that concatenation is successful
cactdf.info()
```

```python
cactdf.fillna('NA')
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

for x in csalesdf["CUSTOMER_NAME"]:
    try:
        b = process.extractOne(x, cactdf["CUSTOMER_NAME"])[0]
        print(f"{x} <-> {b}")
    except TypeError:
        print(f"XXX Problem with '{x}'")
```

```python
best_match_index = lambda x: process.extractOne(x,csalesdf["CUSTOMER_NAME"])[2]
cactdf['BestMatch'] = cactdf["CUSTOMER_NAME"].dropna().map(best_match_index)

csales_cactdf = csalesdf.merge(cactdf, left_index=True, right_on='BestMatch', how='left')
```

```python
def series_statistics(column):
  print('statistics for column: {}'.format(column.name))
  print('The max value in the column: {}'.format(column.max()))
  print('The min value in the column: {}'.format(column.min()))
  print('The mode value in the column: {}'.format(column.mode()))
  print('The median value in the column: {}'.format(column.median()))
  print('The mean of the column: {}'.format(column.mean()))
  print('The std of the column: {}'.format(column.std()))
```
